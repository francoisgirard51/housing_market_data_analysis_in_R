```r
library(dplyr)
```

```r
# Impute missing values with the median for each column
housing_data <- housing_data %>% 
  mutate(
    Crime.Rate = ifelse(is.na(Crime.Rate), median(Crime.Rate, na.rm = TRUE), Crime.Rate),
    Average.Rooms = ifelse(is.na(Average.Rooms), median(Average.Rooms, na.rm = TRUE), Average.Rooms),
    Public.Transport.Access = ifelse(is.na(Public.Transport.Access), median(Public.Transport.Access, na.rm = TRUE), Public.Transport.Access),
    Number.of.Schools = ifelse(is.na(Number.of.Schools), median(Number.of.Schools, na.rm = TRUE), Number.of.Schools),
    Median.Home.Value = ifelse(is.na(Median.Home.Value), median(Median.Home.Value, na.rm = TRUE), Median.Home.Value)
  )

# Checking for any remaining missing values after replacement
sapply(housing_data, function(x) sum(is.na(x)))
```

## Handling missing values

All missing values in the dataset have been successfully replaced with the median of each respective variable. This ensures that the dataset is now complete and ready for further analysis without any missing data. Here’s a summary of missing values per variable after the replacement process:

- **Crime.Rate**: 0 missing values
- **Average.Rooms**: 0 missing values
- **Public.Transport.Access**: 0 missing values
- **Number.of.Schools**: 0 missing values
- **Median.Home.Value**: 0 missing values

With this replacement, the dataset is now fully populated and ready for modeling and further analysis.

# Hypothesis testing

```r
# Simple linear regression: predicting Median.Home.Value using Average.Rooms
linear_model <- lm(Median.Home.Value ~ Average.Rooms, data = housing_data)
summary(linear_model)
```

## Hypothesis Testing - Simple Linear Regression Analysis

We performed a simple linear regression to predict `Median.Home.Value` using `Average.Rooms` as the predictor. Key findings from the model are as follows:

- **Intercept**: 3.7854, suggesting a baseline value when `Average.Rooms` is zero (a theoretical scenario).
- **Coefficient for `Average.Rooms`**: 7.1886, indicating that each additional room is associated with an increase in the median home value by approximately 7.19 units.
- **Significance**: The p-value for `Average.Rooms` is < 2e-16, confirming that it is a statistically significant predictor of `Median.Home.Value`.
- **R-squared**: 0.7588, indicating that about 75.88% of the variability in median home values is explained by the average number of rooms, showing strong predictive power.
- **Residual Standard Error**: 2.698, reflecting the typical deviation of actual values from the predicted values.

The regression analysis supports the hypothesis that homes with more rooms tend to have higher median values, with `Average.Rooms` being a significant and strong predictor of `Median.Home.Value`.

```r
ggplot(housing_data, aes(x = Average.Rooms, y = Median.Home.Value)) +
  geom_point(color = "blue", alpha = 0.5) +
  geom_smooth(method = "lm", color = "red", se = TRUE) +
  labs(title = "Regression of Median Home Value on Average Rooms",
       x = "Average Rooms",
       y = "Median Home Value") +
  theme_minimal()
```

## Final Conclusion of the Housing Price Analysis

This analysis aimed to explore and model factors influencing housing prices in the dataset. The primary findings and methodological steps are as follows:

1. **Data Cleaning and Imputation**:
   - Missing values were imputed using the median of each variable to ensure completeness, with particular attention to `Crime.Rate` and `Average.Rooms`, which had some missing entries.
     
2. **Exploratory Data Analysis**:
   - `Median.Home.Value` showed a roughly normal distribution, indicating a balanced spread of housing prices with fewer extreme values.

3. **Correlation Analysis and Predictor Identification**:
   - The strongest predictor of `Median.Home.Value` was identified as `Average.Rooms`, with an R-squared of 0.76, meaning that this variable explains approximately 76% of the variance in housing prices within the dataset.
     
4. **Linear Regression Model**:
   - A linear regression model confirmed a strong positive, statistically significant relationship between `Average.Rooms` and `Median.Home.Value`. The model suggests that for each additional room, the median home value increases by approximately 7.19 units.
   - Although `Average.Rooms` is a valuable predictor, other variables in the dataset (such as access to public transport and the number of schools) did not significantly influence housing prices.

5. **Limitations and Scope**:
   - This model is specific to the dataset provided and may not generalize to all real estate markets. In practice, other factors like neighborhood quality, accessibility, and school quality can heavily influence housing prices but were not included in this dataset.


In conclusion, while the average number of rooms is a useful predictor for median home values within this dataset, housing prices are inherently multi-dimensional. A more comprehensive analysis would require additional variables to fully capture the complexity of real estate pricing. This model serves as a foundational approach, demonstrating how room count relates to price, but highlighting the need for a holistic approach in predictive modeling.
